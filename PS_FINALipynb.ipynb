```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import random

# Vom determina calea corectă după ce vedem structura directoarelor
# Pentru moment, să folosim o abordare mai simplă

def find_image_directories():
    """Caută directoarele care conțin imagini"""
    base_dir = '/content/Palmprint_Recognition/data'

    # Verifică toate subdirectoarele
    for root, dirs, files in os.walk(base_dir):
        for filename in files:
            if filename.endswith(('.jpg', '.png', '.bmp')):
                print(f"Am găsit imagini în: {root}")
                return root

    return None

def show_sample_images():
    """Afișează câteva imagini de amprente palmare"""
    base_dir = '/content/Palmprint_Recognition/data'

    # Caută imagini recursiv
    image_files = []
    for root, dirs, files in os.walk(base_dir):
        for filename in files:
            if filename.endswith(('.jpg', '.png', '.bmp')):
                image_files.append(os.path.join(root, filename))
                if len(image_files) >= 6:  # Limităm la 6 imagini
                    break
        if len(image_files) >= 6:
            break

    if not image_files:
        print("Nu s-au găsit imagini!")
        return

    # Afișează imaginile găsite
    plt.figure(figsize=(15, 10))
    for i, img_path in enumerate(image_files[:6]):
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        if img is not None:
            plt.subplot(2, 3, i+1)
            plt.imshow(img, cmap='gray')
            plt.title(f"Imagine {i+1}")
            plt.axis('off')

    plt.tight_layout()
    plt.savefig('sample_palmprints.png')
    plt.show()

def plot_accuracy_methods():
    """Afișează un grafic comparativ al metodelor testate"""
    methods = ['SIFT+KNN', 'ResNet18', 'Gaussian+SIFT', 'Gabor+SIFT', 'LBP+SIFT']
    accuracies = [97.31, 83.16, 94.5, 95.2, 93.7]

    plt.figure(figsize=(10, 6))
    bars = plt.bar(methods, accuracies, color=['#3498db', '#e74c3c', '#2ecc71', '#f39c12', '#9b59b6'])

    # Adaugă valorile deasupra barelor
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height + 0.5,
                f'{height:.2f}%', ha='center', va='bottom')

    plt.title('Comparație Metode de Recunoaștere a Amprentelor Palmare')
    plt.ylabel('Acuratețe (%)')
    plt.ylim(0, 105)
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    plt.tight_layout()
    plt.savefig('methods_comparison.png')
    plt.show()

# Execută funcțiile
try:
    # Caută directoarele cu imagini
    image_dir = find_image_directories()
    print(f"Director cu imagini: {image_dir}")

    # Afișează câteva imagini
    show_sample_images()

    # Afișează comparația metodelor
    plot_accuracy_methods()
except Exception as e:
    print(f"Eroare: {str(e)}")
```
